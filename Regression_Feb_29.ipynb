```
#第二个model
from sklearn.preprocessing import PolynomialFeatures
X = np.array(df[['temp','day','weekofyear','hour','windspeed','month']]) #两个变量就不用reshape
y = np.array(df['count'])
X_inter = PolynomialFeatures(degree = 2, include_bias = False).fit_transform(X)
slm = LinearRegression().fit(X=X_inter,y=y)
cv_score2 = cross_val_score(slm,X_inter,y,cv =5, scoring = 'neg_root_mean_squared_error')
print(-cv_score2.mean())
print(cv_score2)
```
